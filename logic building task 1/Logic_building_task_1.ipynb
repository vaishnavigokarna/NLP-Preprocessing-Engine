# TASK 1: CONCEPTUAL ANSWERS
# 1. What is the difference between "Love" and "love" in NLP?

In NLP, “Love” and “love” can be treated as two different words if the text is not converted into lowercase. This happens because computers read uppercase and lowercase letters differently. For example, “Love” may appear at the beginning of a sentence, while “love” may appear in the middle. Even though both have the same meaning, the model may count them as separate tokens. That is why lowercasing is an important preprocessing step in NLP, as it helps make the text more consistent and reduces unnecessary duplication of words.

# 2. What happens if stopwords are not removed?

If stopwords are not removed, the text may contain many common words like “is”, “the”, “and”, “of”, and “in”, which usually do not add much meaning to the analysis. Because of this, the dataset becomes noisier and larger, and the model may focus less on the important words. It can also increase processing time and reduce the effectiveness of tasks like keyword extraction or text classification. However, stopwords are not always useless, because in some cases they are necessary to preserve the meaning of a sentence.

# 3. Provide two real-world scenarios where removing stopwords can be harmful.

Removing stopwords can be harmful in tasks like sentiment analysis and question-answering systems. For example, in sentiment analysis, the sentence “I do not like this product” becomes incorrect if the word “not” is removed, because the meaning changes completely. Similarly, in question-answering or search systems, a sentence like “What is the capital of India?” may lose its proper meaning if important connecting words are removed. In such cases, stopwords help preserve the original intent and context of the sentence.

# 4. Explain the difference between stemming and lemmatization.

Stemming and lemmatization are both techniques used to reduce words to their base form, but they work differently. Stemming simply cuts off the ending parts of words, sometimes producing incomplete or non-dictionary words. For example, “studies” may become “studi”. Lemmatization is more advanced because it converts a word into its correct dictionary root form using language rules. For example, “studies” becomes “study” and “running” becomes “run”. In simple terms, stemming is faster but less accurate, while lemmatization is more meaningful and accurate.

# Task 2: Build Advanced Preprocessing Function

```python
# Importing required libraries

import re
import pandas as pd
from collections import Counter
```

```python
# Preprocessing function

def preprocess_text(text):
    # Handle empty or invalid input
    if text is None or not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", " ", text)

    # Remove emails
    text = re.sub(r"\S+@\S+\.\S+", " ", text)

    # Remove emojis and special characters
    text = re.sub(r"[^\w\s']", " ", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)

    # Remove numbers
    text = re.sub(r"\d+", " ", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short words but keep meaningful ones
    keep_words = {"no", "not", "ok", "in", "on", "at", "to"}
    cleaned_tokens = [word for word in tokens if len(word) > 2 or word in keep_words]

    # Join tokens into cleaned sentence
    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

#Task 3: Stress Testing

```python
# Sample input sentences

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
# Processing all sentences

results = []

for sentence in sample_inputs:
    tokens, clean_sentence = preprocess_text(sentence)

    results.append({
        "Original Text": sentence,
        "Cleaned Tokens": tokens,
        "Cleaned Sentence": clean_sentence
    })

results
```

```python
# Displaying processed output

for i, row in enumerate(results, 1):
    print(f"\nSentence {i}")
    print("Original Text   :", row["Original Text"])
    print("Cleaned Tokens  :", row["Cleaned Tokens"])
    print("Cleaned Sentence:", row["Cleaned Sentence"])
    print("-" * 80)
```

```python
# Creating dataframe for processed results

df_results = pd.DataFrame(results)
df_results
```

#Task 4: Token Analytics

```python
# Token analytics

analytics = []

for row in results:
    tokens = row["Cleaned Tokens"]

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_token_length = round(sum(len(token) for token in tokens) / total_tokens, 2)
    else:
        avg_token_length = 0

    analytics.append({
        "Original Text": row["Original Text"],
        "Total Tokens": total_tokens,
        "Unique Tokens": unique_tokens,
        "Average Token Length": avg_token_length
    })

df_analytics = pd.DataFrame(analytics)
df_analytics
```

```python
# Finding most noisy and most meaningful sentence

noise_scores = []

for row in results:
    original_word_count = len(row["Original Text"].split())
    cleaned_word_count = len(row["Cleaned Tokens"])
    removed_words = original_word_count - cleaned_word_count

    noise_scores.append({
        "Sentence": row["Original Text"],
        "Removed Words": removed_words,
        "Remaining Tokens": cleaned_word_count
    })

most_noisy = max(noise_scores, key=lambda x: x["Removed Words"])
most_meaningful = max(noise_scores, key=lambda x: x["Remaining Tokens"])

print("Sentence with the most noise:")
print(most_noisy["Sentence"])

print("\nSentence that retained the most meaningful tokens after cleaning:")
print(most_meaningful["Sentence"])
```

# Task 6:Build Full Pipeline

```python
# Full pipeline function

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.append(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
# Running full pipeline

pipeline_output = full_pipeline(sample_inputs)
pipeline_output
```

```python
# Displaying full pipeline output

print("TOKENS:")
for i, tokens in enumerate(pipeline_output["tokens"], 1):
    print(f"Sentence {i}: {tokens}")

print("\nCLEAN SENTENCES:")
for i, sentence in enumerate(pipeline_output["clean_sentences"], 1):
    print(f"Sentence {i}: {sentence}")
```
